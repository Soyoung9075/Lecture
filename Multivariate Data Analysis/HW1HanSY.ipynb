```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기

auto = pd.read_fwf("auto.dat", header = None, widths = [19,4,6,3,2,2,4,5,3,5,4,3,4,4], 
                   na_values='.')
auto.columns = ['model','origin', 'price', 'mpg', 'rep78', 'rep77', 'hroom', 'rseat', 
                'trunk', 'weight', 'length', 'turn', 'displa', 'gratio']
```

```python
# 필요 변수만 별도로 데이터 생성

var_list = ["price", "mpg", "hroom", "rseat", "weight", "length", "gratio"]
x = auto[var_list]
x
```

```python
# Sample mean vector

def mean(data):
    ones = pd.DataFrame({'ones': np.ones((len(data)))}) # one vector 생성
    mean_data = ones.transpose().dot(data).transpose() / len(data)
    return mean_data

mean(x)
```

```python
# Sample variance-covariance

def cov(data):
    n = len(data)
    I = np.identity(len(data))
    ones = np.ones(n)
    J = ones.reshape((n,1)).dot(ones.reshape((1,n)))
    m = (I- (1/(n))*J) @ data
    cov = (m.transpose().dot(m)) / (n-1)
    return cov

cov(x)
```

```python
# Sample correlation

def corr(data):
    cov_matrix = cov(data)
    var_list = np.diag(cov_matrix).reshape((-1,1))
    var_matrix = np.sqrt(var_list @ var_list.T)
    corr = cov_matrix / var_matrix
    return corr

corr(x)
```

## 2. 
Write a Python code to test the significance of correlation between two input
variables. Using Applicant data, test the significance of correlations between the
variable pairs (LC,SC), (SC,SMS), and (LC,SMS), respectively. (P-value
계산목적으로 Python패키지 혹은 내장함수 사용 가능)

```python
# 데이터 불러오기

applicant = pd.read_csv('applican.dat', header = None, delim_whitespace=True)
applicant.columns = ['ID','FL','APP','AA','LA','SC','LC','HON','SMS','EXP',
                     'DRV','AMB','GSP','POT','KJ','SUIT']
applicant.head(5)
```

```python
# LC, SC, SMS 변수만 별도로 데이터 프레임 생성
applicant_2 = applicant[['LC', 'SC', 'SMS']] 

# 문제 1에서 만든 correlation 함수 사용하여 correlation matrix 생성
corr_app = corr(applicant_2)
corr_app
```

```python
# correlation between each pair

print("Correlation between LC and SC : {}".format(corr_app.iloc[(0,1)]))
print("Correlation between SC and SMS : {}".format(corr_app.iloc[(2,1)]))
print("Correlation between LC and SMS : {}".format(corr_app.iloc[(0,2)]))
```

```python
# Hypothesis Testing

n = len(applicant)

# Correlations for each pair
corr_1 = corr_app.iloc[(0,1)] # LC and SC
corr_2 = corr_app.iloc[(2,1)] # SC and SMS
corr_3 =corr_app.iloc[(0,2)] # LC and SMS

# 각 correlation에 대한 Test statistics 생성
t_stat_1 = (corr_1*np.sqrt(n-2)) / (np.sqrt(1-corr_1**2))
t_stat_2 = (corr_2*np.sqrt(n-2)) / (np.sqrt(1-corr_2**2))
t_stat_3 = (corr_3*np.sqrt(n-2)) / (np.sqrt(1-corr_3**2))

print("1. Test statistic for corr_1 : {}".format(t_stat_1))
print("2. Test statistic for corr_2 : {}".format(t_stat_2))
print("3. Test statistic for corr_3 : {}".format(t_stat_3))
```

```python
# 각 test statistic에 대한 p-value 구하기 

import scipy.stats as stats

print(("1. P-value {}".format((1- stats.t.cdf(t_stat_1, df = n-2))*2)))
print(("2. P-value {}".format((1- stats.t.cdf(t_stat_2, df = n-2))*2)))
print(("3. P-value {}".format((1- stats.t.cdf(t_stat_3, df = n-2))*2)))
```

# 3.
Write a Python code to get the confidence interval for correlation between two
input variables. Using Applicant data, find the confidence interval for correlations
between the variable pairs (LC,SC), (SC,SMS), and (LC,SMS), respectively. Use
both the Fisher’s method and the Ruben’s method, then compare.

```python
# Fisher's Method

def ci_fisher(rho, alpha, n):
    alpha = alpha
    u = stats.norm.ppf(1-alpha/2, loc=0, scale=1)
    upper = np.tanh(np.arctanh(rho) + u / np.sqrt(n-3))
    lower = np.tanh(np.arctanh(rho) - u /np.sqrt(n-3))
    return lower, upper

n = len(applicant)
print("Confidence Interval for corr between LC and SC : {}".format(ci_fisher(corr_1, 0.05,n)))
print("Confidence Interval for corr between SC and SMS : {}".format(ci_fisher(corr_2, 0.05,n)))
print("Confidence Interval for corr between LC and SMS : {}".format(ci_fisher(corr_3, 0.05,n)))
```

```python
# Ruben's Method

def ci_ruben(rho, alpha, n):
    alpha = alpha
    u = stats.norm.ppf(1-alpha/2, loc=0, scale=1)
    r_star = rho / np.sqrt(1-rho**2)
    a = 2*n - 3 - u**2
    b = r_star*np.sqrt((2*n-3)*(2*n-5))
    c = (2*n-5-(u**2))*(r_star**2)-2*(u**2)
    y_1 = (2*b - np.sqrt(4*(b**2)-4*a*c)) / (2*a)
    y_2 = (2*b + np.sqrt(4*(b**2)-4*a*c)) / (2*a)
    lower = y_1 / (np.sqrt(1 + y_1**2))
    upper = y_2 / (np.sqrt(1 + y_2**2))
    return lower, upper

n = len(applicant)
print("Confidence Interval for corr between LC and SC : {}".format(ci_ruben(corr_1, 0.05,n)))
print("Confidence Interval for corr between SC and SMS : {}".format(ci_ruben(corr_2, 0.05,n)))
print("Confidence Interval for corr between LC and SMS : {}".format(ci_ruben(corr_3, 0.05,n)))
```

# 4.
Impute missing values in ‘rep78’ variable in ‘auto.dat’ using Python. Use a
regression method using other variables of your choice. Calculate means and
standard deviations of ‘rep78’ variable before and after the imputation. (이
문제에서는 파이썬 패키지 혹은 내장함수를 사용해도 됨.)

```python
# auto 데이터셋의 NA 현황 파악 

auto.isna().sum()
```

```python
# rep77의 결측치를 나머지 결측치가 없는 변수들로 예측

import warnings
warnings.filterwarnings('ignore')

import statsmodels.api as sm

# 기존 dataframe에서 'rep78'에 NA가 있는 행 삭제
auto_rev = auto.dropna(subset=['rep77'], axis = 0)

# Input과 target 데이터 셋 나누기
# Input에서 rep78, rep77, model, origin만 제외하고 나머지 변수 모두 포함
dfX = auto_rev[auto.columns.difference(['rep78', 'rep77', 'model', 'origin'])]
dfy = auto_rev[['rep77']]

# Linear Regression 피팅
model = sm.OLS(dfy, dfX)
result = model.fit()

# 위 모델로'rep77'의 NA 예측
na_index_list = auto.index[auto['rep77'].isna()].tolist()
X_new = auto.iloc[na_index_list][auto.columns.difference(['rep78', 'rep77', 'model', 'origin']).tolist()]

y_pred = result.predict(X_new)
y_pred
```

```python
# rep77 NA를 예측값으로 대체

df_pred = pd.DataFrame(y_pred, columns = ['rep77'])
auto_fill = pd.concat([auto_rev[['rep77']], df_pred], axis = 0).sort_index()
auto_new = pd.concat([auto[auto.columns.difference(['rep77'])], auto_fill], axis =1)
auto_new
```

```python
# rep78 결측치 imputation

# 'rep78'에 NA가 있는 행 삭제
auto_rev = auto_new.dropna(subset=['rep78'], axis = 0) 

# Input과 target 데이터 셋 나누기
# Input에 rep78, model, origin만 제외하고 나머지 변수 모두 포함
dfX = auto_rev[auto.columns.difference(['rep78', 'model', 'origin'])] 
dfy = auto_rev[['rep78']]

# Linear Regression 피팅
model = sm.OLS(dfy, dfX)
result = model.fit()
print(result.summary())
```

```python
# 위 모델로 'rep78' NA 예측
na_index_list = auto_new.index[auto_new['rep78'].isna()].tolist()
X_new = auto_new.iloc[na_index_list][auto_new.columns.difference(['rep78','model', 'origin']).tolist()]

y_pred = result.predict(X_new)
y_pred
```

```python
# Imputation 전과 후의 rep78 평균 및 표준편차 계산

df_pred = pd.DataFrame(y_pred, columns = ['rep78'])
auto_fill = pd.concat([auto_rev[['rep78']], df_pred], axis = 0)

print("Before imputation \n mean: {:.5f} / standard deviation : {:.5f}".format(auto['rep78'].mean(), auto['rep78'].std()))
print("After imputation \n mean: {:.5f} / standard deviation : {:.5f}".format(auto_fill['rep78'].mean(), auto_fill['rep78'].std()))
```
